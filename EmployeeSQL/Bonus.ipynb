```python
# sqlalchemy for DB to Python
from sqlalchemy import create_engine
# password for sqlalchemy
from config import password
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'postgresql://postgres:' + password + '@localhost:5432/SQL_Homework'
engine = create_engine(url)
connection = engine.connect()
```

```python
employees_df = pd.read_sql_table("employees",con=engine)
employees_df
```

```python
departments_df = pd.read_sql_table("departments",con=engine)
departments_df
```

```python
dept_emp_df = pd.read_sql_table("dept_emp",con=engine)
dept_emp_df
```

```python
dept_manager_df = pd.read_sql_table("dept_manager",con=engine)
dept_manager_df
```

```python
salary_df = pd.read_sql_table("salary",con=engine)
salary_df
```

```python
titles_df = pd.read_sql_table("titles",con=engine)
titles_df
```

```python
sal_bins = [0,25000,50000,75000,100000,125000,150000]
sal_titles = ['0-25K','25K-50K','50K-75K','75K-100K','100K-125K','125K-150K']
```

```python
plt.figure()
salary_df.hist(column='salary')
plt.title("Common Salary Ranges")
plt.xlabel('Salary')
plt.ylabel('Count')
```

```python
sal_title_merge = salary_df.merge(titles_df,on='employeeid')
sal_title_df = sal_title_merge.groupby('title')['salary'].mean()
sal_title_df
```

```python
sal_title_bar = sal_title_df.plot(kind='bar')
sal_title_bar.set_title('Average Salary by Title')
sal_title_bar.set_xlabel('Title')
sal_title_bar.set_ylabel('Average Salary ($)')
```
